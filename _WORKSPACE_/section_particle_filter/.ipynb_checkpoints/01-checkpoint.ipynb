```python
from robot import *
from scipy.stats import multivariate_normal
import random
import copy
import matplotlib.pyplot as plt
```

```python
class Particle:
    def __init__(self, init_pose, weight):
        self.pose = init_pose
        self.weight = weight
        
    def motion_update(self, nu, omega, time, noise_rate_pdf):
        ns = noise_rate_pdf.rvs()
        noised_nu = nu + ns[0]*math.sqrt(abs(nu)/time) + ns[1]*math.sqrt(abs(omega)/time)
        noised_omega = omega + ns[2]*math.sqrt(abs(nu)/time) + ns[3]*math.sqrt(abs(omega)/time)
        self.pose = IdealRobot.state_transition(noised_nu, noised_omega, time, self.pose)
```

```python
class Mcl:
    def __init__(self, envmap, init_pose, num, motion_noise_stds={'nn':0.1, 'no':0.001, 'on':0.02, 'oo':0.001}):
        self.particles = [Particle(init_pose, 1.0/num) for i in range(num)]
        self.map = envmap     
        
        v = motion_noise_stds
        c = np.diag([v['nn']**2, v['no']**2, v['on']**2, v['oo']**2])
        self.motion_noise_rate_pdf = multivariate_normal(cov=c)
        
    def motion_update(self, nu, omega, time):
        for p in self.particles: p.motion_update(nu, omega, time, self.motion_noise_rate_pdf)
        
    def dist(self, elems, time): 
        self.xs = [p.pose[0] for p in self.particles]
        self.ys = [p.pose[1] for p in self.particles] 
        if int(time)%10 ==0 and time - int(time) == 0.0:elems[int(time)] = {'x': self.xs, 'y': self.ys}
```

```python
class EstimationAgent(Agent):
    def __init__(self, time_interval, nu, omega, estimator):
        super().__init__(nu, omega)
        self.estimator = estimator
        self.time_interval = time_interval
        
        self.prev_nu = 0.0
        self.prev_omega = 0.0
        
    def decision(self, observation=None):
        self.estimator.motion_update(self.prev_nu, self.prev_omega, self.time_interval)
        self.prev_nu, self.prev_omega = self.nu, self.omega
        return self.nu, self.omega
    
    def dist(self, elems, time):
        self.estimator.dist(elems, time)
```

```python
def straight():
    time_interval = 0.1
    world = World(100, time_interval, debug=False)
    m = Map()

    ### ロボットの作成 ###
    initial_pose = np.array([np.random.normal(0.0,1.0),np.random.normal(0.0,2.0),np.random.normal(0.0,5.0)/180*math.pi]).T
    estimator = Mcl(m, initial_pose, 100)
    a = EstimationAgent(time_interval, 1.0, 0.0, estimator)
    r = IdealRobot(initial_pose, agent=a)
    world.append(r)

    world.dist()
    return world.elems


particle_data={}
for _ in range(10):
    raw_data = straight()
    for j in raw_data:
        if j not in particle_data.keys():
            particle_data[j] = {}
            particle_data[j]['x'] = raw_data[j]['x'] 
            particle_data[j]['y'] = raw_data[j]['y'] 
        else:
            particle_data[j]['x'] += raw_data[j]['x'] 
            particle_data[j]['y'] += raw_data[j]['y'] 
```

```python
fig = plt.figure(figsize=(10,10)) 
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 17
ax = fig.add_subplot(111)
ax.set_aspect('equal')   
ax.set_xlabel("Y[m]",fontsize=16)                 
ax.set_ylabel("X[m]",fontsize=16) 
ax.grid(True)

for i in particle_data:
    x,y = particle_data[i]['x'], particle_data[i]['y']
    if i == 100:
        ax.scatter(y, x, label=str(i)+'[sec]', color='navy')
    else:
        ax.scatter(y,x, label=str(i)+'[sec]')
ax.legend(fontsize=15)
plt.savefig('01.png')
plt.show()
```
